```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown


def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
#Start searching with google gemini
from google.colab import userdata
GOOGLE_API_KEY = userdata.get('GoogleApiKey')
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
def ask_question(name):
  ques=input("Hey! " + name + " How can I help?: ")
  return ques
```

```python
def classify_ques(ques):
  webSearch = False
  deviceList = ['alarm','reminder','message','call']
  personalList = ['who are you?','who created you?']

  device = False
  for i in deviceList:
    if i in ques.lower():
      device = True

  if device:
    print("This question is related to Device things, which is not supported in our assistant.")


  personalQues= False
  for i in personalList:
    if i in ques.lower():
      personalQues = True

  if personalQues:
    print("Hey!")

  if device:
    webSearch = False
  elif personalQues:
    webSearch = False
  else:
    webSearch = True

  return webSearch

classify_ques('who are you?')
```

```python
model = genai.GenerativeModel('gemini-pro')
response = model.generate_content("What is the meaning of life?")
to_markdown(response.text)
```

```python
def askGemini(ques):
  response = model.generate_content(ques)
  to_markdown(response.text)
  return response.text
```

```python
from IPython.display import Audio
def speak(answer):
  tts=gTTS(answer)
  tts.save('audio.mp3')
  display(Audio('audio.mp3',autoplay = False))
```

```python
#Flow
anyOtherQues = 'y'
name = ''
while anyOtherQues == 'y':
  if name == '':
    name = input("Hey! What is your name?: ")
  q = ask_question(name)
  go_ahead = classify_ques(q)

  textReply = ''
  if go_ahead == True:
    textReply = askGemini(q)
    speak(textReply)
    print(textReply)

  anyOtherQues = input("Do you have any other question?")
```
